```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os
import dlib
```

# 얼굴 인식 및 스티커 붙이기 객체 정의

- CameraApp(background_path, sticker_path, model_path): 배경(얼굴이 있는) 사진 경로, 스티커 이미지 경로, 사전학습된 모델 경로 입력



- show_img(): 불러온 배경 및 스티커 이미지를 보고 싶은 경우 사용하는 메서드



- get_face_size(): 얼굴 인식 후, 얼굴 사이즈를 반환하는 메서드



- find_landmarks(): 얼굴 인식 및 눈 코 입 등 랜드마크를 찾아 반환하는 메서드



- add_sticker(): 얼굴에 사진을 붙이고, 합성된 이미지를 보여주는 메서드. 얼굴 위치 및 크기에 따라 사진의 크기를 조절. 



```python
class CameraApp:
    def __init__(self, background_path, sticker_path, model_path='camera_sticker/models/shape_predictor_68_face_landmarks.dat'):
        self.background_path = background_path
        self.sticker_path = sticker_path 
        self.background = self.load_img(background_path)
        self.model = dlib.shape_predictor(model_path)
        self.face_size = self.get_face_size(self.background)
        self.sticker = self.load_img(sticker_path)
        self.landmarks = self.find_landmarks()
        
    def load_img(self, path):
        img = cv2.imread(path)
        ratio = img.shape[0] / img.shape[1]
        img = cv2.resize(img, (640, int(ratio * 640)))
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        return img
    
    
    def show_img(self):
        plt.imshow(self.background)
        plt.show()
        plt.imshow(self.sticker)
        plt.show()
    
    
    def get_face_size(self, img):
        face_detector = dlib.get_frontal_face_detector()
        face_points = face_detector(img, 1)
        if not face_points:
            plt.imshow(img)
            raise ValueError('얼굴 인식이 되지 않습니다. 다른 사진을 넣어주세요.')
        bounding_box = face_points[0]
        l = bounding_box.left()
        t = bounding_box.top()
        r = bounding_box.right()
        b = bounding_box.bottom()
        return (r-l, b-t)
    
    
    def find_landmarks(self):
        detector_hog = dlib.get_frontal_face_detector()
        dlib_rects = detector_hog(self.background, 1)
        landmarks = []
        for dlib_rect in dlib_rects:
            points = self.model(self.background, dlib_rect)
            points = list(map(lambda p: (p.x, p.y), points.parts()))
            return points
    
    
    def add_sticker(self, w_ratio, h_ratio, size_ratio=1):
        # w_ratio: 코 기준 - 얼굴사이즈 비율로 '좌우' 이동
        # h_ratio: 코 기준 - 얼굴사이즈 비율로 '상하' 이동
        # size_ratio: 스티커 크기 비율
        new_img = self.background.copy()
        nose_x, nose_y = self.landmarks[30]
        face_h, face_w = self.face_size
        new_sticker_size = (int(face_h * size_ratio), int(face_w * size_ratio))
        sticker = cv2.resize(self.sticker, new_sticker_size)
        
        h, w, _ = sticker.shape
        
        new_x = nose_x + int(w * w_ratio)
        new_y = nose_y + int(h * h_ratio)
        h0 = max(0, new_y-h//2)
        h1 = min(new_img.shape[0], new_y+h//2)
        
        w0 = max(0, new_x-w//2)
        w1 = min(new_img.shape[1], new_x+w//2)
    
        
        
        sticker_area = new_img[h0:h1, w0:w1]
        if not h0 and not w0:
            sticker = sticker[h0-h1:, w0-w1:]
        elif not h0:
            sticker = sticker[h0-h1:, :w1-w0]
        elif not w0:
            sticker = sticker[:h1-h0, w0-w1:]
        else:
            sticker = sticker[:h1-h0, :w1-w0]
        sticker_area = sticker_area[:sticker.shape[0], :sticker.shape[1]]
        new_img[h0:h1, w0:w1] = np.where(sticker==0, sticker_area, sticker).astype(np.uint8)
        
        plt.imshow(new_img)
```

### 회전 개선

```python
class NewCameraApp(CameraApp):
    def __init__(self, background_path, sticker_path, model_path='camera_sticker/models/shape_predictor_68_face_landmarks.dat'):
        self.rotate_cnt = 0
        super().__init__(background_path, sticker_path, model_path)
        
    
    def get_face_size(self, img):
        face_detector = dlib.get_frontal_face_detector()
        for _ in range(4):
            self.rotate_cnt += 1
            img = np.rot90(img)
            face_points = face_detector(img, 1)
            if face_points:
                self.background = img
                break
        if not face_points:
            plt.imshow(img)
            raise ValueError('얼굴 인식이 되지 않습니다. 다른 사진을 넣어주세요.')
        bounding_box = face_points[0]
        l = bounding_box.left()
        t = bounding_box.top()
        r = bounding_box.right()
        b = bounding_box.bottom()
        return (r-l, b-t)
    
    def add_sticker(self, w_ratio, h_ratio, size_ratio=1):
        new_img = self.background.copy()
        nose_x, nose_y = self.landmarks[30]
        face_h, face_w = self.face_size
        new_sticker_size = (int(face_h * size_ratio), int(face_w * size_ratio))
        sticker = cv2.resize(self.sticker, new_sticker_size)
        
        h, w, _ = sticker.shape
        
        new_x = nose_x + int(w * w_ratio)
        new_y = nose_y + int(h * h_ratio)
        h0 = max(0, new_y-h//2)
        h1 = min(new_img.shape[0], new_y+h//2)
        
        w0 = max(0, new_x-w//2)
        w1 = min(new_img.shape[1], new_x+w//2)
    
        sticker_area = new_img[h0:h1, w0:w1]
        if not h0 and not w0:
            sticker = sticker[h0-h1:, w0-w1:]
        elif not h0:
            sticker = sticker[h0-h1:, :w1-w0]
        elif not w0:
            sticker = sticker[:h1-h0, w0-w1:]
        else:
            sticker = sticker[:h1-h0, :w1-w0]
        sticker_area = sticker_area[:sticker.shape[0], :sticker.shape[1]]
        new_img[h0:h1, w0:w1] = np.where(sticker==0, sticker_area, sticker).astype(np.uint8)
        for _ in range(4-self.rotate_cnt):
            new_img = np.rot90(new_img)
        plt.imshow(new_img)
```

### 배경이 밝거나 어두운 사진은 인식이 잘되는데, 얼굴이 밝거나 어두운 사진은 인식이 왜 잘 안될까?

우리가 사용한 모델은 HOG 필터를 이용한 모델

HOG 필터는 빛의 밝기의 변화도를 이용한 모델인데, 얼굴자체가 밝거나 어두울 경우, 빛의 밝기의 변화도가 비슷비슷해 이미지의 특성을 파악하기 어려워진다. 따라서 얼굴 인식에 실패하게 된다.

1. CNN 을 활용한 필터를 사용해보자.
2. 직접 붙이고 싶은 지점을 선택해서 스티커를 붙이자.

# 해결

## CNN 모델 제작

CNN 모델 제작을 위한 데이터셋  
https://www.kaggle.com/c/facial-keypoints-detection/data

```python
import pandas as pd
import numpy as np
from keras.layers.advanced_activations import LeakyReLU, ReLU
from keras.models import Sequential, Model
from keras.layers import Activation, Convolution2D, MaxPooling2D, BatchNormalization, Flatten, Dense, Dropout, Conv2D,MaxPool2D, ZeroPadding2D
```

```python
train_df = pd.read_csv("camera_sticker/images/train/training.csv")
test_df = pd.read_csv("camera_sticker/images/test/test.csv")
```

```python
train_df = train_df.dropna()
test_df = test_df.dropna()
```

```python
len(train_df)
```

### 데이터 셋 제작
- 예측하고자 하는 위치만을 남기자.
- 양쪽 눈, 코, 아랫입술만 남기고, 나머지 좌표는 삭제한다.

```python
drop_cols = ['left_eye_inner_corner_x', 'left_eye_inner_corner_y', 'left_eye_outer_corner_x', 'left_eye_outer_corner_y',
            'right_eye_inner_corner_x', 'right_eye_inner_corner_y', 'right_eye_outer_corner_x', 'right_eye_outer_corner_y',
            'left_eyebrow_inner_end_x', 'left_eyebrow_inner_end_y', 'left_eyebrow_outer_end_x', 'left_eyebrow_outer_end_y',
             'right_eyebrow_inner_end_x', 'right_eyebrow_inner_end_y', 'right_eyebrow_outer_end_x', 'right_eyebrow_outer_end_y',
            'mouth_left_corner_x','mouth_left_corner_y','mouth_right_corner_x','mouth_right_corner_y',
             'mouth_center_top_lip_x','mouth_center_top_lip_y']
train_df = train_df.drop(drop_cols, axis=1)
```

```python
x = train_df['Image'].str.split(' ', expand=True)
x = np.array(x, dtype=int)
x = x.reshape(-1, 96, 96, 1)
x = x / 255.0
print(x.shape)
```

```python
# 왼쪽눈 x, y,  오른쪽눈 x, y,  코 x,y, 아래입술 중앙 x,y
y = train_df.iloc[:, :-1]
y = np.array(y, dtype=float)
print(y.shape)
```

```python
test = test_df['Image'].str.split(' ', expand=True)
test = np.array(test, dtype=int)
test = test.reshape(-1, 96, 96, 1)
test = test / 255.0
print(test.shape)
```

### 모델 정의

```python
model = Sequential()

model.add(Convolution2D(32, (3,3), padding='same', input_shape=(96,96,1)))
model.add(ReLU())
model.add(MaxPool2D(pool_size=(2, 2)))

model.add(BatchNormalization())
model.add(Convolution2D(64, (3,3), padding='same'))
model.add(ReLU())
model.add(MaxPool2D(pool_size=(2, 2)))
          
model.add(BatchNormalization())
model.add(Convolution2D(128, (3,3), padding='same'))
model.add(ReLU())
model.add(MaxPool2D(pool_size=(2, 2)))
          
model.add(BatchNormalization())
model.add(Convolution2D(256, (3,3), padding='same'))
model.add(ReLU())
model.add(MaxPool2D(pool_size=(2, 2)))
          
model.add(BatchNormalization())
model.add(Convolution2D(512, (3,3), padding='same'))
model.add(ReLU())
model.add(MaxPool2D(pool_size=(2, 2)))
          
model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(8))
model.summary()
```

```python
model.compile(optimizer='adam', loss='mse', metrics=['mse', 'mae'])
```

### 모델 훈련

```python
model.fit(x, y, epochs=30, batch_size=32, validation_split=0.2)
```

### 예측 및 결과

```python
pred = model.predict(test)
```

### CNN 모델을, 앞의 CameraApp에 사용할 수 있을까?
- **No**
- CNN 모델은 얼굴정면을 학습했음. CameraApp에 쓰이는 사진들은 각양각색의 상황에서 얼굴을 탐지하고, 랜드마크를 탐지해야함.
- 어떻게 해결?
- 우선 얼굴 bounding box를 탐지하는 모델을 통해 얼굴을 영역을 예측하고, 위 모델을 사용하거나
- 새로운 데이터셋으로 학습 진행해야 하거나,
- Augmentation을 통해 해결한다.

## 직접 붙이기 (번외)
스마트폰에서 사용하는 스티커앱의 경우도, 사용자가 직접 스티커를 붙일 때가 있다.  
번외로, 주피터 노트북에서 사용자 입력을 받아 스티커를 붙일 위치를 지정하는 방법에 대해 알아보자.

```python
import cv2
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

```python
def load_img(path):
    img = cv2.imread(path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img
```

```python
def add_sticker(background, sticker, loc):
    fig = plt.figure(figsize=(10, 10))
    loc = loc[-4:]
    xs = sorted([x[0] for x in loc])
    ys = sorted([x[1] for x in loc])
    w0, w1 = int(xs[0]), int(xs[-1])
    h0, h1 = int(ys[0]), int(ys[-1])
    w, h = w1-w0, h1-h0
    sticker = cv2.resize(sticker, (w, h))
    h, w, _ = sticker.shape
    
    sticker_area = background[h0:h1, w0:w1]
    background[h0:h1, w0:w1] = np.where(sticker==0, sticker_area, sticker).astype(np.uint8)

    plt.imshow(background)
    plt.show()
```

```python
%matplotlib notebook

fig = plt.figure(figsize=(7, 7))

img = load_img('camera_sticker/images/dark image.jpg')
sticker = load_img('camera_sticker/images/king.png')

imgplot = plt.imshow(img)
loc = []
def onclick(event):
    global loc
    ix, iy = event.xdata, event.ydata
    loc.append((ix, iy))
    if len(loc) == 4:
        get_ipython().magic('matplotlib inline')

cid = fig.canvas.mpl_connect('button_press_event', onclick)

plt.show()
```

```python
add_sticker(img, sticker, loc)
```
